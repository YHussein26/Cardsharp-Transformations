```python
#To brighten the red, while darkening the green colors, move the values of the color closer to 255; 
#To darken the green, use(old color value/255)^2*255

#To brighten the image even more than accomplished by the square-root transformation, you have  to increase
#the color value (initial one)
#To darken all color values above 128 but brighten all color values below 128, you have to
#decrease the color value (initial) up to 128, then increase the color values (initial) from 128
```

```python
img = plt.imread('cardsharps.jpg')
print(img.shape)
plt.clf()
plt.imshow(img[:,:,0],cmap='Blues',vmin=50,vmax=100)
plt.colorbar()
plt.show()
```

```python
import numpy as np 
import PIL

img2 = img.copy()
img2[:,:,:] = [255*np.sqrt(x/255) for x in img2[:,:,:]]
plt.clf()
plt.imshow(img2)
from PIL import Image
im = Image.open('cardsharps.jpg')
im.rotate(85).show()
```
